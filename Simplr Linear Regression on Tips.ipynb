```python
# Import All Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

```

```python
# Load the data into a Pandas DataFrame
data = pd.read_excel('Tips.xlsx')

# Check the first few rows
data.head()
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Identify columns to drop
columns_to_drop = [col for col in data.columns if 'Unnamed' in col]

# Drop the identified columns
data = data.drop(columns_to_drop, axis=1)

# Print the cleaned dataset
data.head(5)
```

```python

# Visualize the relationship between total bill and tip
sns.scatterplot(x='total_bill', y='tip', data=data)
plt.show()
```

```python
# Assuming you cleaned the data in a separate step
X = data[['total_bill']]  # Update the feature matrix if necessary
y = data['tip']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train) 
```

```python
# Make predictions on the testing set
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error:', mse)
print('R-squared:', r2)
```
